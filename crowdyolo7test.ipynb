##  Area Measurement ZT30

```python
import math

# Given parameters
distance = 20  # distance from camera to subject in meters
horizontal_fov = 58.1  # horizontal field of view in degrees
diagonal_fov = 65.4  # diagonal field of view in degrees

# Calculating half angle and aspect ratio
half_angle = math.radians(horizontal_fov / 2)
aspect_ratio = math.tan(math.radians(diagonal_fov / 2)) / math.tan(half_angle)

# Calculating width and height of visible area
visible_width = 2 * distance * math.tan(half_angle)
visible_height = visible_width / aspect_ratio

# Calculating total visible area
visible_area = visible_width * visible_height

print(f"Visible Width: {visible_width} meters")
print(f"Visible Height: {visible_height} meters")
print(f"Total Visible Area: {visible_area} square meters")
```

```python
import math

def calculate_area_dimensions(distance, dfov, hfov):
    # Calculate vertical FOV
    vfov = 2 * math.atan(math.tan(math.radians(dfov / 2)) * 3 / 4)
    
    # Calculate width and height
    width = 2 * distance * math.tan(math.radians(hfov / 2))
    height = 2 * distance * math.tan(math.radians(vfov / 2))
    
    return width, height

# Given values
distance = 50
dfov = 65.4
hfov = 58.1

# Calculate area dimensions
width, height = calculate_area_dimensions(distance, dfov, hfov)

# Print the results
print(f"Area Width: {width:.2f} meters")
print(f"Area Height: {height:.2f} meters")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math

def calculate_area_dimensions(distance, dfov, hfov):
    # Calculate vertical FOV
    vfov = 2 * math.atan(math.tan(math.radians(dfov / 2)) * 3 / 4)

    # Calculate width and height
    width = 2 * distance * math.tan(math.radians(hfov / 2))
    height = 2 * distance * math.tan(vfov / 2)
    
    return width, height

# Given values
distance = 20
dfov = 65.4
hfov = 58.1

# Calculate area dimensions
width, height = calculate_area_dimensions(distance, dfov, hfov)

# Create a plot
fig, ax = plt.subplots()

# Add a rectangle to the plot
rectangle = patches.Rectangle((0, 0), width, height, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rectangle)

# Set axis labels
ax.set_xlabel('Width (meters)')
ax.set_ylabel('Height (meters)')

# Set aspect ratio to be equal
ax.set_aspect('equal', adjustable='box')

# Show the plot
plt.show()
```
